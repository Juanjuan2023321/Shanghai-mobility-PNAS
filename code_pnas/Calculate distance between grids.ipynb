```python
from geopy.distance import geodesic
from math import sin, asin, cos, radians, fabs, sqrt
import math
import numpy as np
import pandas as pd
from tqdm import tqdm_notebook as tqdm
```

```python
def get_distance(lat1, long1, lat2, long2):
    EARTH_RADIUS = 6372422
    lat1 = lat1*math.pi/180
    long1 = long1*math.pi/180
    lat2 = lat2*math.pi/180
    long2 = long2*math.pi/180
    diff_long = long1-long2
    diff_lat = lat1-lat2
    a = np.sin(diff_lat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(diff_long/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    d = EARTH_RADIUS*c
    return d
```

```python
grid_df = pd.read_excel('/SH_travel/pc/grid_loc.xlsx')
```

```python
res = []
for i in tqdm(range(1, len(grid_df))):
    lat1, long1 = np.array([grid_df['lat'][i-1]]*(len(grid_df)-i)), np.array([grid_df['long'][i-1]]*(len(grid_df)-i))
    lat2, long2 = np.array(grid_df['lat'][i:]), np.array(grid_df['long'][i:])
    d = get_distance(lat1, long1, lat2, long2)
    res.append([0]*i+list(d))
res.append([0]*len(grid_df))

res = pd.DataFrame(res)
res.columns = list(grid_df['grid'])
res.index = list(grid_df['grid'])
res.head(10)
```

```python
res.to_csv('/SH_travel/pc/grid_distance_od.csv')
```
